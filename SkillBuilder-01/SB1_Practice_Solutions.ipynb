![](https://media.giphy.com/media/cFLLnExjELn7a/giphy.gif)


f-strings, short for "functional" strings, in Python really make it easier to print information to the screen (also referred to as the "console"). This might seem like a silly thing to do for simple code, but it's *extremely* beneficial when doing calculations and checking if they're right before running code that will affect tens of thousands of rows of data!

Get in the habit of printing to understand what your code is doing since this a great way to learn the nuances of Python 🐍.

**Your task is to write the following sentence using the variables defined below.**

For example, if your name was `Lauren` and your favorite food was `vegemite` your f-string should print the following:

```
Hi, my name is Lauren. I came here to eat vegemite and write Python code... and it looks like I am all out of vegemite!
```



````python
# Complete the code below
my_name = 'Lauren'
favorite_food = 'vegemite'

# write your f-string here
print(f'Hi, my name is {my_name}. I came here to eat {favorite_food} and write Python code... and it looks like I am all out of {favorite_food}!')

# you can also save your f-string to a variable then print the variable
statement = f'Hi, my name is {my_name}. I came here to eat {favorite_food} and write Python code... and it looks like I am all out of {favorite_food}!'
print(statement)
````
