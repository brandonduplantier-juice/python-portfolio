## LevelUp - Visualizing Total Jobs by District

A choropleth map is an effective tool for visualizing data distribution across geographical areas. It helps to uncover patterns, trends, and clusters that might not be apparent in a table or a bar graph.

In this Task, you'll create a choropleth map to visualize total jobs by district.


You will need two key components:
1. **Geographical Data**: This consists of the boundaries (shapes or polygons) for the regions you want to visualize—such as districts, states, or countries. This data is typically stored in formats like GeoJSON or TopoJSON.
2. **Job Data**: A dataset that includes the total number of jobs in each district.


We are fortunate to have a GeoJSON file that contains the district boundaries for Nepal. You simply need to run the provided cell to load this data into your notebook.

````python
# Run this cell!
import requests
import json

geojson_url = "https://raw.githubusercontent.com/mesaugat/geoJSON-Nepal/37f73c5cac420447d1d58b51e1a890bb608bf48a/nepal-districts-new.geojson"

response = requests.get(geojson_url)
geojson_data = response.json()
````

