### Basic Game Flow
The basic steps in a Tic-Tac-Toe program are:

**Display the Board**: Show the empty Tic-Tac-Toe board.

**Player Input**: Ask Player 'X' to pick a position to place their 'X'.

**Update the Board**: Reflect the player's choice on the board.

**Check for Outcome**: See if Player 'X' wins or if it's a tie.

**Switch Turns**: Repeat steps 2-4 for Player 'O'.

## Tic-Tac-Toe Functions

The three functions below have already been coded!
- `display_board(board)` displays the game board to the player.
- `is_winner(board, symbol)` determines if a given symbol, X or O, has won the game at any given instant.
- `is_board_full(board)` determines if the board is full, i.e., the game ends in a tie.

There is **no need** to modify these functions! But remember to run the cells so that code you do write can use them.

### 👉 The `display_board` Function
This function will show a grid! The Tic-Tac-Toe player will select the choices counting left-to-right from the top-left to the bottom-right of the grid.

```{python}
1 | 2 | 3
--+---+--
4 | 5 | 6
--+---+--
7 | 8 | 9
```

The `board` variable in this notebook will be a **list**.

````python
# RUN THIS CELL - DO NOT MODIFY
def display_board(board):
    '''
    Function to display the board to the player. Player will select a number
    between 1 - 9 to select where to make their next move.
    e.g.
    1 | 2 | 3
    --+---+--
    4 | 5 | 6
    --+---+--
    7 | 8 | 9
    '''
    print(f"{board[0]} | {board[1]} | {board[2]}")
    print("--+---+--")
    print(f"{board[3]} | {board[4]} | {board[5]}")
    print("--+---+--")
    print(f"{board[6]} | {board[7]} | {board[8]}")
````

##### EXAMPLE USAGE

````python
# An example board
example_board = ['X',' ','O',
                 ' ','O',' ',
                 ' ','X',' ']

# Display board
display_board(example_board)
````

### 👉 The `is_winner` Function
This function is used to determine if a given symbol, X or O, has won the game at any given instant. Output will return `True` if the symbol wins the game, `False` otherwise.

````python
# RUN THIS CELL - DO NOT MODIFY
def is_winner(board, symbol):
    '''
    Function to determine if a given symbol, X or O, has won the game
    at any given instant.

    Output will return True if the symbol wins the game, False otherwise.
    '''
    return (
        (board[0] == board[1] == board[2] == symbol) or
        (board[3] == board[4] == board[5] == symbol) or
        (board[6] == board[7] == board[8] == symbol) or
        (board[0] == board[3] == board[6] == symbol) or
        (board[1] == board[4] == board[7] == symbol) or
        (board[2] == board[5] == board[8] == symbol) or
        (board[0] == board[4] == board[8] == symbol) or
        (board[2] == board[4] == board[6] == symbol)
    )
````

##### EXAMPLE USAGE

````python
# An example board
example_board = ['X','O','O',
                 'X','O','X',
                 'X','X','O']

# Is there a winner?
is_winner(example_board, 'X')
````

### 👉 The `is_board_full` Function
This function is used to determine if the board is full, i.e., the game ends in a tie.  Output will return `True` if the board is full, `False` otherwise.

````python
# RUN THIS CELL - DO NOT MODIFY
def is_board_full(board):
    '''
    Function to determine if the board is full, i.e., the game ends in a tie.
    '''
    # Return True only if *all* cells in the board are not equal to a space (' ')
    return all(cell != ' ' for cell in board)
````

##### EXAMPLE USAGE

````python
# An example board
example_board = ['X',' ','O',
                 ' ','O',' ',
                 ' ','X',' ']

# Is the board full?
is_board_full(example_board)
````

## Play the Tic-Tac-Toe game!
The `play_game()` this function will synthesize all the elements of the game and allow the (human) player to play against the computer in a Tic Tac Toe death match.

No need to modify this function! So long as `get_player_move` and `get_computer_move` are properly coded, `play_game` will allow you to, well, play the game!

````python
def play_game():
    '''
    Wrapper function to play the game. Uses helper functions defined above.
    '''
    
    # Create a blank board with 9 empty spaces
    board = [' ' for _ in range(9)]

    # Set a flag to control the game loop
    continue_playing = True

    while continue_playing:
        # Human player always goes first (symbol: 'X')

        # Display the current board
        display_board(board)

        # Get the player's move and update the board
        board = get_player_move(board)

        # Check if player 'X' has won
        if is_winner(board, 'X'):
            display_board(board)
            print("Player X wins!")
            break

        # Check for a tie (i.e., the board is full with no winner)
        if is_board_full(board):
            display_board(board)
            print("It's a tie!")
            break

        # Computer's turn (symbol: 'O')

        # Get the computer's move and update the board
        board = get_computer_move(board)

        # Check if computer 'O' has won
        if is_winner(board, 'O'):
            display_board(board)
            print("Computer wins!")
            break

        # Check again for a tie
        if is_board_full(board):
            display_board(board)
            print("It's a tie!")
            break
````

````python
play_game()
````
