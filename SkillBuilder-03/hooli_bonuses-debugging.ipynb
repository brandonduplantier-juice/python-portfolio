# Debugging Hooli Bonuses

Remember, Hooli wants to distribute year-end bonuses based on employee tenure. 

*   Employees who have been with the company for less than 5 years will receive 5% of their total salary as a bonus 
*   Employees who have been with the company for more than 5 years, but less than 10 years will receive 10% of their total salary as a bonus 
*   And employees who have been with the company for more than 10 years will receive 15% of their total salary as a bonus

Is $50 million enough money to say 'thanks' to their employees?

```python
# load data 
import sys
sys.path.append('./files')
from hooli import hooli_employees
```

```python
hooli_employees['000001']
```

```python
#iterate over every employee at Hooli
bonuses = 0
for id in hooli_employees:
    #store employee tenure and employee salary
    employee_tenure = hooli_employees[id][1]
    employee_salary = hooli_employees[id][2]
    
    #compute how much money the employee earns in a bonus 
    if employee_tenure < 5:
        bonuses += employee_salary*0.05
    elif employee_tenure < 10:
        bonuses += employee_salary*0.1
    else:
        bonuses += employee_salary*0.15
```

```python
# print whether or not $50 million is enough
print(f'The total amount of money needed for bonuses is: ${bonuses}')
if bonuses <= 50000000:
    print('$50,000,000 is enough money to thank Hooli employees!')
else:
    print('Uh oh! Hooli will need a larger thank-you budget.')
```
