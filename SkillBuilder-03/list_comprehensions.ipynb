## List Comprehensions for Transformations

Below, is a list of daily high temperatures (in Celsius) in Paris, France from September 2019.

```python
temperatures = [22, 22, 23, 24, 18, 20, 19,
                18, 18, 22, 23, 23, 23, 25,
                26, 26, 21, 19, 19, 23, 27,
                22, 20, 20, 18, 20, 20, 19,
                20, 18]
```

Create a new list of temperatures where these highs have been converted into units of Fahrenheit.

> F = C * 9 / 5 + 32

```python
# create a list to store converted temperatures
temps_f = []

# convert each temperature and add to new list
for temp in temperatures:
    temps_f.append(temp * 9 / 5 + 32)

# after loop, print out converted temperatures
print(temps_f)
```

## List Comprehensions for Filtering

```python
# load the data 
import sys
sys.path.append('./files')
from ws_winners import ws_winners
```

Stored in the `ws_winners` dictionary are World Series years and their winning teams. Keys are the years, and values are the winning teams. 

How many World Series titles have been won by the New York Yankees?

```python
# set up a counter variable
win_years = []

# loop over the dictionary years
for year in ws_winners:
    # add tally if winner was the Yankees
    if ws_winners[year] == 'New York Yankees':
        win_years.append(year)

# print the final win count
print(len(win_years))
```
