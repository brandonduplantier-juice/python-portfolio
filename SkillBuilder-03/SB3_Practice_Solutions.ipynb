### Effort 1 - Time to Party!

Imagine you're having a house party and you invited several of your friends. Unfortunately, word got out to people you don't know and don't want at your house.

![](https://media.giphy.com/media/KSCYjxdjm0yA0/giphy.gif)

In this effort we're going to write code to simulate every single person that comes and knocks on your door. If that person is in your invite list you'll greet them by their name and welcome them to your home with the message `Hey <Name>! Glad you could make it. Come on in!`

If that person isn't in your invite list, you will instead tell them `I'm sorry, you have the wrong house.`


````python
door_knockers = ['Richard', 'Ana', 'Sandeep', 'Jamie', 'Lauren', 'Bobby', 'Paul', 'Bobby', 'Scott', 'Jamal', 'Diane', 'Umesh', 'Carlos', 'Ali', 'Juan']
invited_list = ['Bobby', 'Umesh', 'Sandeep', 'Jamal', 'Lauren', 'Ana', 'Paul', 'Fitzgerald', 'German', 'Juan', 'Jamie']

# write your for loop below
# your code should have 1 if statement and 1 else statement inside the loop

for knocker in door_knockers:
    if knocker in invited_list:
        print(f'Hey {knocker}! Glad you could make it. Come on in!')

    else:
        print("I'm sorry, you have the wrong house.")
````

````python
# dictionary of friends and their favorite foods
friends_dict = {
    "Charles":"pizza",
    "Jamie":"korean BBQ",
    "Umesh":"lamb Vindaloo",
    "Omar":"pizza",
    "Lauren":"breakfast cereal",
    "Lys":"shrimp fettucine Alfredo",
    "Marshall":"pizza",
    "Robert":"arepas",
    "Orly":"gelato",
    "Delaney":"ramen",
    "Morgan":"donuts",
    "Mike":"pizza",
    "Matt":"chicken wings",
    "John":"tacos",
    "Eric":"bagels",
    "Chris":"cheeseburger quesadillas", 
    "Isaac":"ham & cheese sandwiches",
    "Becky":"croquetas",
    "Cody":"pizza"
}

# write code below

# track pizzas
pizza_count = 0

for friend in friends_dict:
    food = friends_dict[friend]

    # print message to friend
    print(f'Hi {friend}! thanks for bringing {food}, looks delicious!')

    # check if pizza
    if food == 'pizza':
        pizza_count += 1

print(f'Wow! What are we going to do with {pizza_count} pizzas!?!?')
````
