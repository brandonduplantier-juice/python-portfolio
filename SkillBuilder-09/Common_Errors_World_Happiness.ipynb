```python
# Columns to rename in the 2015 DF
new_col_names_2015 = {
    "Happiness Score":"happiness_2015", 
    "Economy (GDP per Capita)":"per_capita_gdp_2015",
    "Health (Life Expectancy)":"life_expectancy_2015"
}

# Columns to rename in the 2019 DF
new_col_names_2019 = {
    "Score":"happiness_2019", 
    "GDP per capita":"per_capita_gdp_2019", 
    "Healthy life expectancy":"life_expectancy_2019"}
```

```python
# Rename the happiness score columns to include the year
df_2015 = df_2015.rename(columns=new_col_names_2015)
df_2019 = df_2019.rename(columns=new_col_names_2019)
```

```python
# When merging DFs, need left_on and right_on arguments to specify columns 
# that share common values
joined_df = 
```

```python
# left_on and right_on arguments for pd.merge() are specified
# Join the two DataFrames with an inner join
joined_df = pd.merge(df_2015, df_2019, 
                     left_on="Country", 
                     right_on="Country or region", 
                     how="inner")
```

```python
# Preview the joined DataFrame
joined_df.head()
```

```python
# Drop the extra column
joined_df = joined_df.drop(columns=["Country or region"])
```

```python
# Create a new column of the happiness score change from '15 to '19
joined_df["happiness_change_2015_to_19"] = joined_df["happiness_2019"] - joined_df["happiness_2015"]
```

```python
# Sort the DF by biggest drop in happiness
joined_df.sort_values(by="happiness_change_2015_to_19").head(10)
```

```python
# Sort the DF by biggest increase in happiness
joined_df.sort_values(by="happiness_change_2015_to_19", ascending=False).head(10)
```
