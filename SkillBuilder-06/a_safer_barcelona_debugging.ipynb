```python
# import libraries
import pandas as pd
import plotly.express as px

accidents = pd.read_csv('barcelona_accidents.csv')
```

## Visualization #1:
Using the `barcelona_accidents.csv` data create a visualization to display the number of accidents, by day of the week.

```python
# Histogram of daily accidents 
px.histogram(accidents, x='weekday',
            title='Barcelona Car Accidents by Day of the Week')
```
